```python
import re
import random
```

```python
# Expresiones regulares para saludos y consultas
saludos = re.compile(r'^(buen(as (noch|tard)e|os dia)s|saludos|buenas|h(e(l{2}o|y)|o{1,2}la|i))$', re.IGNORECASE)
clima = re.compile(r'[^a-z]*(soleado|estado|grados|nieve|clima|l{2}(over|uvia)|frio|sol)', re.IGNORECASE)
stock = re.compile(r'[^a-z]*(cotizacion|dow jones|acciones|mercado|baja |alta)', re.IGNORECASE)
afirmacion = re.compile(r'\b(si|claro|ok|por\s?supuesto)\b', re.IGNORECASE)
despedida = re.compile(r'\b(adios|chao|hasta\s?(luego|pronto)|nos\s?vemos)\b', re.IGNORECASE)
negativa = re.compile(r'\b(no|adios|gracias)\b', re.IGNORECASE)
```

```python
# Respuestas predeterminadas
respuesta_clima = [
    'El clima está nublado con temperatura de 20°C.',
    'Se espera una posibilidad de lluvia para el resto del día.',
    'Hoy el sol está brillando con una temperatura de 25°C.',
    'El clima es mayormente soleado con algunas nubes dispersas.'
]
respuesta_stock = [
    'Las acciones están a la baja, podría ser un buen momento para comprar.',
    'El mercado está estable, pero se prevé una subida en el futuro cercano.',
    'Las acciones están subiendo, es una buena oportunidad para vender.',
    'El mercado está inestable, se recomienda seguir monitoreando.'
]
respuesta_default = [
    'Lo siento, no entiendo tu pregunta. ',
    'No estoy seguro de cómo responder a eso.',
    'No entiendo bien lo que quieres decir. '
]
```

```python
# Función para responder la pregunta del usuario
def responder_pregunta(pregunta):

    if clima.search(pregunta):
        return random.choice(respuesta_clima)
    elif stock.search(pregunta):
        return random.choice(respuesta_stock)
    else:
        return random.choice(respuesta_default)
```

```python
# Función principal del chatbot
def chatbot():

    while True:
        inicio = input("Hola!, soy chatbot ")
        if saludos.search(inicio):
            print("Te puedo comentar acerca del clima o el mercado de acciones.")
            pregunta = input("¿Cuál es tu consulta? ")
            respuesta = responder_pregunta(pregunta.lower())
            print(respuesta)

            otra_pregunta = input("¿Te gustaría preguntar algo más? (si/no) ")
            if afirmacion.search(otra_pregunta.lower()):
                pregunta2 = input("¿Cuál es tu siguiente consulta? ")
                respuesta = responder_pregunta(pregunta2.lower())
                print(respuesta)
                print("Solo puedo responder 2 veces, un gusto Adios")
                break

            if negativa.search(otra_pregunta.lower()) or despedida.search(otra_pregunta.lower()):
                print("¡Adiós! Que tengas un buen día.")
                break
            else:
                print("No es una afirmación, espero haberte ayudado.")
                print("¡Hasta luego!")
                break
        else:
            print(f"mmm, no reconozco '{inicio}'. Por favor, usa un saludo como 'hola'")
```

```python
if __name__ == "__main__":
    chatbot()
```
