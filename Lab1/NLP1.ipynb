```python
import re
import random
```

```python
saludo = "(?<!xyz)h"

clima = re.compile(r'\b(?:frio|clima|sol|lluvia)\b')
stock = re.compile(r'\b(?:acciones|cotiza|subida)\b')

afirmacion = re.compile(r'\b(?:si|claro|ok)\b')
```

```python
respuesta_clima = ['clima nublado', 'temperatura de 20Â°C', 'posibilidad de lluvia']
respuesta_stock = ['acciones a la baja', 'tiempo de comprar acciones']
```

```python
while True:
    inicio = input("Hola usuario! ")
    if not re.match(saludo, inicio):
        print ("Error! No reconozco tu saludo")
    else:
        print("Te puedo comentar acerca del clima o stock market")
        pregunta = input("Cual es tu consulta? ")
        if clima.search(pregunta.lower()):
          print(random.choice(respuesta_clima))
        elif stock.search(pregunta.lower()):
          print(random.choice(respuesta_stock))
        else:
          print("No entiendo tu pregunta")
        otra_pregunta = input("Te gustaria preguntar algo mas? ")
        if afirmacion.search(otra_pregunta.lower()):
          pregunta2 = input("Cual es tu siguiente consulta? ")
          if clima.search(pregunta2.lower()):
            print(random.choice(respuesta_clima))
          elif stock.search(pregunta2.lower()):
            print(random.choice(respuesta_stock))
          else:
            print("No entiendo tu pregunta")
          print("Adios")
        else:
          print("Adios")
        break
```
